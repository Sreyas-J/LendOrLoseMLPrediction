```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = [
    "NaiveBayesClassifier",
    "LogisticRegression",
    "KNN",
    "DT",
    "RF",
    "Adaboost",
    "XGB",
    "GB"
]

data = {
    "NaiveBayesClassifier": [0.6788, 0.6821, None, None, None, None],
    "LogisticRegression": [0.67, 0.68, None, None, None, None],
    "KNN": [0.8839, 0.9380, 0.6607, 0.8838, 0.9321, 0.9321],
    "DT": [0.9390, 0.6441, 0.8822, 0.8822, 0.9390, 0.9390],
    "RF": [0.8858, 0.99, 0.6802, 0.8857, 0.99, 0.9389],
    "Adaboost": [0.8854, 0.6808, None, 0.8854, 0.6808, None],
    "XGB": [0.8862, 0.7114, 0.6839, 0.8862, 0.7114, 0.7114],
    "GB": [0.8861, 0.694, 0.6816, 0.8816, 0.694, 0.9389]
}

data_types = ["normalized", "normalized_oversampled", "normalized_undersampled", "RAW", "oversampled", "undersampled"]

# Plot
bar_width = 0.1
x_labels = np.arange(len(models))

# Create the bar graph
plt.figure(figsize=(15, 8))

colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']

for i, dtype in enumerate(data_types):
    values = [data[model][i] if data[model][i] is not None else 0 for model in models]
    plt.bar(x_labels + i * bar_width, values, width=bar_width, label=dtype, color=colors[i])

# Customizations
plt.xticks(x_labels + bar_width * (len(data_types) - 1) / 2, models, rotation=45, fontsize=10)
plt.ylabel("Validation Accuracy", fontsize=12)
plt.xlabel("Models", fontsize=12)
plt.title("Model Accuracy by Data Preprocessing Type", fontsize=14)
plt.legend(title="Data Type", fontsize=10)
plt.ylim(0, 1.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
